```python
# Code generated by Data Wrangler for PySpark DataFrame

def clean_data(df):
    # Filter rows based on column: 'PO Line Reference'
    df = df.filter(~df['PO Line Reference'].eqNullSafe("-1"))
    # Replace missing values with "<na>" in columns: 'PO Line Reference', 'Reporting Code Description - 1st' and 7 other columns
    df = df.fillna(value="<na>", subset=['PO Line Reference', 'Reporting Code Description - 1st', 'Reporting Code Description - 2nd', 'Reporting Code Description - 3rd', 'Vendor Name', 'Interested Users', 'Material Type', 'PO Line Notes', 'PO Line Type Name'])
    return df

df_clean = clean_data(df)
display(df_clean)
```
