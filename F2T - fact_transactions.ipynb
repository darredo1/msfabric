```python
df = spark.read.format("csv").option("header","true").load("abfss://abfs path.csv")
# df now is a Spark DataFrame containing CSV data from "Files/Alma Analytics/Fund Transactions/Fact - Fund Transactions.csv".
display(df)
```

```python
# Code generated by Data Wrangler for PySpark DataFrame

from pyspark.sql import types as T

def clean_data(df):
    # Filter rows based on column: 'PO Line Reference'
    df = df.filter(~df['PO Line Reference'].eqNullSafe("-1"))
    # Change column type to datetime64[ns] for column: 'Transaction Date'
    df = df.withColumn('Transaction Date', df['Transaction Date'].cast(T.TimestampType()))
    # Change column type to float64 for column: 'Line Price'
    df = df.withColumn('Line Price', df['Line Price'].cast(T.DoubleType()))
    # Replace missing values with "<na>" in columns: 'PO Line Reference', 'Transaction Id' and 3 other columns
    df = df.fillna(value="<na>", subset=['PO Line Reference', 'Transaction Id', 'MMS Id', 'Fund Code', 'Invoice Line-Unique Identifier'])
    return df

df_clean = clean_data(df)
display(df_clean)
```

```python
df_clean.write.format("delta").mode('overwrite').option("delta.columnMapping.mode", "name").saveAsTable("fact_transactions")

df_clean.printSchema()
```
